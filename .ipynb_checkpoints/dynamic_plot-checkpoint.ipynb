```python
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import geopandas as gpd
```

```python
geo_df=gpd.read_file('C:/Users/user/Desktop/test_plotly/13_26_agosto_ZERO.shp')
```

```python
list(geo_df)
```

```python
fig = go.Figure(data=go.Scattergeo(
        lat=geo_df.geometry.y,
        lon=geo_df.geometry.x,
        mode = 'markers',
        marker = dict(
        size = 5,
        opacity = 0.8,
        reversescale = True,
        autocolorscale = False,
        symbol = 'square',
        line = dict(
        width=0,
        color='rgba(102, 102, 102)'),
        colorscale = 'Reds',
        cmin = 0,
        color = geo_df['13_08_2020'],
        cmax = geo_df['13_08_2020'].max(),
        colorbar_title=""
        )))

fig.update_geos(
    center=dict(lon=8.937, lat=45.932),
    lataxis_range=[45.55,46.35], lonaxis_range=[7.8,9]
)

fig.update_layout(
    width=500,
    height=300,
    autosize=True,
    margin=dict(t=0, b=0, l=0, r=0),
    template="plotly_white",
)

fig.update_layout(
    updatemenus=[
        dict(
            buttons=list([
                dict(
                    args=["type", "surface"],
                    label="3D Surface",
                    method="restyle"
                ),
                dict(
                    args=["type", "heatmap"],
                    label="Heatmap",
                    method="restyle"
                )
            ]),
            direction="down",
            pad={"r": 10, "t": 10},
            showactive=True,
            x=0.1,
            xanchor="left",
            y=1.1,
            yanchor="top"
        ),
    ]
)
```

```python
fig1 = go.Scattermapbox(
        lat=geo_df.geometry.y,
        lon=geo_df.geometry.x,
        mode='markers',
        marker=go.scattermapbox.Marker(
            size=5,
            color='rgb(255, 0, 0)',
            opacity=0.7
        ),
        text=geo_df.quadkey,
        hoverinfo='text'
    )
)


fig2 = go.Scattermapbox(
        lat=geo_df.geometry.y,
        lon=geo_df.geometry.x,
        mode='markers',
        marker=go.scattermapbox.Marker(
            size=5,
            color='rgb(255, 0, 0)',
            opacity=0.7
        ),
        text=geo_df.quadkey,
        hoverinfo='text'
    )
)
             


fig.update_layout(
    title='Nuclear Waste Sites on Campus',
    width=500,
    height=300,
    autosize=True,
    hovermode='closest',
    showlegend=False,
    mapbox=dict(
        bearing=0,
        center=dict(lon=8.937, lat=45.932),
        pitch=0,
        zoom=7.5,
        style='light'
    ),
)

fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})

fig.show()
```
